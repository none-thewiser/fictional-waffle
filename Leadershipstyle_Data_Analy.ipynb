## Data Preparation and Cleaning

Here first I'm importing pandas as pd 
Then I will use read_csv to read the file which I have uploaded from my computer to this jupiter notebook

```python
import pandas as pd
```

```python
df = pd.read_csv('2021_population.csv')
```

```python
df
```

Now I'm going to rename one of the columns from the above dataset.
To do so, I'm using rename method from pandas. 
Dataframe.rename(columns= {'old_value':'new_value}, inplace=True)

```python
df.rename(columns= {'2021_last_updated': '2021_updated_population'}, inplace=True)
```

```python
country_growth_rate_df = df[['iso_code','country', '2021_updated_population', 'growth_rate', 'world_%']]
```

```python
country_growth_rate_df
```

Here I'm going to use one more dataset, 'Democracy-Dictatorship_Index'

```python
df1 = pd.read_csv('Democracy-Dictatorship_Index.csv')
```

```python
df1
```

I'm selecting only two columns from the above dataset and saving it into a new dataframe.

```python
Demo_Dict_df = df1[['Regime', 'Type']]
```

```python
Demo_Dict_df
```

Here I'm renaming Regime to iso-code

```python
Demo_Dict_df.rename(columns ={'Regime':'iso_code'}, inplace=True)
```

```python
Demo_Dict_df
```

Here I'm going to merge both the dataframes with having iso_code in common to form one large dataframe.

```python
Combined_df = country_growth_rate_df.merge(Demo_Dict_df, on='iso_code')
```

```python
Combined_df
```

Now, I'm using one more dataset GDP per capita in csv format.
This is the list of countries of the world by gross domestic product per capita, i.e., the purchasing power parity value of all final goods and services produced within a country in a given year. 

```python
df3 = pd.read_csv('GDP per capita.csv')
```

```python
df3
```

This is another data set Life expectancy.
Life expectancy is a statistical measure of the average time an organism is expected to live, based on its year of birth, its current age, and other demographic factors including gender. 

```python
df4 = pd.read_csv('Life expectancy.csv')
```

```python
df4
```

This is one more dataset, Meat consumption.
Meat consumption by kg per capita as published by the Food and Agriculture Organisation of the United Nations(FAO)

```python
df5 = pd.read_csv('Meat consumption.csv')
```

```python
df5
```

This is another dataset of sex-ratio

It represents the human sex ratio over a country, by donating the ratio of males for each female in a population.

```python
df7 = pd.read_csv('Sex-ratio.csv')
```

```python
df7
```

This is the dataset of Urbanization
It represents the population shift from rural to urban areas.

```python
df8 = pd.read_csv('Urbanization rate.csv')
```

```python
df8
```

Here again I'm using rename method in order to have a common base among all the datasets. 

```python
Combined_df.rename(columns={'iso_code':'ISO-code'}, inplace=True)
```

```python
Combined_df
```

Now I'm going to merge each dataset into one, step by step. 

```python
first_merge = Combined_df.merge(df4, on='ISO-code')
```

```python
first_merge
```

```python
sec_merge = first_merge.merge(df3, on='ISO-code') 
```

```python
sec_merge
```

```python
th_merge = sec_merge.merge(df5, on='ISO-code')
```

```python
th_merge
```

```python
f_merge = th_merge.merge(df7, on='ISO-code')
```

```python
f_merge
```

```python
final_dataset = f_merge.merge(df8, on='ISO-code')
```

Now we have the final dataset which contains all the columns of different datasets. 

```python
final_dataset
```

Now I'm going to delete some columns that we don't need or are in extras. 

```python
final_dataset.drop(columns='Country_y', inplace=True)
```

```python
final_dataset.drop(columns='Country_x', inplace=True)
```

Here I have got rid of Country_x and Country_y as they were repeatitive in nature with the help of drop method.

```python
final_dataset
```

### Q1: Mention the top 10 countries with higher growth rate!

```python
more_grow_rate_country = final_dataset.sort_values('growth_rate', ascending=False).head(10)
```

```python
more_grow_rate_country.set_index('ISO-code', inplace=True)
```

```python
more_grow_rate_country.rename(columns={'world_%': 'world_percentage'}, inplace=True)
```

```python
more_grow_rate_country
```

Above we have the list of top 20 countries whose growth rates are higher. I have calculated it by using sort_values and putting ascending to False which gave us numbers in descending order. And to get top 20 results I have simply passed 20 in the head. Head gives us values from top and tail gives us value from below.  

### Q2: Name the countries who have the highest life expectancy?

```python
more_LE_country.set_index('ISO-code', inplace=True)
```

```python
more_LE_country
```
