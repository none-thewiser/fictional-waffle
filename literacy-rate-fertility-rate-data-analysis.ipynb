# Part-1 Countries as per Descending order

```python
top_descend_order_country_tl
```

```python
descend_tl_df = top_descend_order_country_tl.fillna({'2010': top_descend_order_country_tl['2010'].mean(), '2020':top_descend_order_country_tl['2020'].mean(), 
                                                    '2011': top_descend_order_country_tl['2011'].mean(), '2012': top_descend_order_country_tl['2012'].mean(), 
                                                    '2013': top_descend_order_country_tl['2013'].mean(), '2014': top_descend_order_country_tl['2014'].mean(), 
                                                    '2015': top_descend_order_country_tl['2015'].mean(), '2016': top_descend_order_country_tl['2016'].mean(),
                                                    '2017': top_descend_order_country_tl['2017'].mean(), '2018': top_descend_order_country_tl['2018'].mean(), 
                                                    '2019': top_descend_order_country_tl['2019'].mean(), })

```

```python
descend_tl_df
```

```python
descend_tl_df.set_index('Country Name')
```

```python
descend_tl_df['total_literacy_deviation_2020_2010'] = descend_tl_df['2020']-descend_tl_df['2010']
```

```python
descend_tl_df
```

```python
descend_tl_df.drop(columns='total_literacy_rate_difference_2020_2010')
```

```python
top_descend_order_country_f
```

```python
descend_f_df = top_descend_order_country_f.fillna(method='pad', axis=1)
```

```python
descend_f_df
```

```python
descend_f_df.set_index('Country Name')
```

```python
descend_f_df['fertility_deviation_2020_2010'] = descend_f_df['2020']-descend_f_df['2010']
```

```python
descend_f_df
```

```python
c_df = descend_f_df.merge(descend_tl_df, on='Country Name')
```

```python
c_df
```

```python
c_df.rename(columns={'Country Name': 'country_name'}, inplace=True)
```

```python
c_df
```

```python
import matplotlib.pyplot as plt
```

```python
import seaborn as sns
```

```python
sns.set_style('darkgrid')
```

```python
difference_over_the_decade_df = c_df[['country_name', 'total_literacy_deviation_2020_2010', 'fertility_deviation_2020_2010']]
```

```python
difference_over_the_decade_df.set_index('country_name')
```

```python
plt.figure(figsize=(16, 6))
plt.plot(c_df['2010_x'],c_df.country_name, 's-b')
plt.plot(c_df['2020_x'],c_df.country_name, 'o--r')

plt.ylabel('Country Name')
plt.xlabel('Rate')

plt.title("Fertility Rate Of Countries(Descending Order)")
plt.legend(['2010 Fertility Rate', '2020 Fertility Rate']);
```

```python
plt.figure(figsize=(16, 6))
plt.plot(c_df['2010_y'],c_df.country_name, 's-b')
plt.plot(c_df['2020_y'],c_df.country_name, 'o--r')

plt.ylabel('Country Name')
plt.xlabel('Rate')

plt.title('Countries Adult Literacy Rate')
plt.legend(['2010 Literacy Rate', '2020 Literacy Rate']);
```

```python
plt.figure(figsize=(12, 6))
plt.title('Difference over the decade(2020-2010)')
sns.scatterplot(x='total_literacy_deviation_2020_2010', 
                y='fertility_deviation_2020_2010', 
                hue='country_name',
                s=100,
                data=difference_over_the_decade_df);
```

# Data Visualisation and Analysis

## Total fertility rate represents the number of children that would be born to a woman if she were to live to the end of her childbearing years and bear children in accordance with age-specific fertility rates of the specified year.

## Adult literacy rate is the percentage of people ages 15 and above who can both read and write with understanding a short simple statement about their everyday life.

# Conclusion

## * First we have taken the top 10 countries from below(or in descending order) and analyzed that over the years as we move from 2010 to 2020, Fertility rate has declined where as the literacy rate has increased. The same thing we have found in another data frame where countries was taken from top(or in ascending order). 

## ** The difference taken from 2010 to 2020 also resulted in the same interpretation that over the years the literacy rate has increased and fertility rate declined. 

# Hence, We can say that, "Countries with More Literacy Rate Tend to have Low Fertility Rate.
